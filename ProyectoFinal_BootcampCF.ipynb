## Reporte 3.- Operacion Promedio

###¿Qué es esto?

Una de las métricas más importantes que podemos monitorear y en algún punto precedir es el monto por operación o por ticket.

Esto es el calculo promedio por cada operación realizada en cierto tiempo, por lo pronto estamos trabajando considerando todo el espectro del tiempo.

### Un primer reporte

```python
#Primero agrupamos por numero de operación por tienda y período de tiempo
rpt_sales = df_main.groupby(by=['Op_MonthYear', 'Store', 'Op_Num'], as_index=False).agg({'Amount':'sum'})

#Creamos una columna adicional copiada del monto para poder reutilizar un calculo adicional
rpt_sales['Amount_Bis']=rpt_sales['Amount']

#Ahora con los datos únicos de operación los sumamos en monto y calculamos el monto promedio en el tiempo y por tienda
rpt_sales_op = rpt_sales.groupby(by=['Op_MonthYear', 'Store'], as_index=False).agg(
        {'Amount':['sum'], 'Amount_Bis':['mean']}
       )

#Para poder graficar los datos renombramos las columnas que se crearon en el proceso de agrupado y calculado
rpt_sales_op.columns = rpt_sales_op.columns.droplevel(level=1)
rpt_sales_op.rename(columns={'Amount':'Sales', 'Amount_Bis':'AvgOP'}, inplace = True)

rpt_sales_op

#Quitamos datos que se salen del bloque principal de datos
#rpt_sales_op = rpt_sales_op[(rpt_sales_op['AvgOP']<4000)].sort_values(['AvgOP'])
#rpt_sales_op
```

```python
alt.Chart(rpt_sales_op).mark_point().encode(
    x='Op_MonthYear',
    y='Sales',
    size='AvgOP',
    color= alt.Color('Store', title='Tiendas',
              scale = alt.Scale(range=['#003790', '#33559d', '#4c74ab', '#6195b9', '#72b7c9', '#80dbdb', '#ffd3bf', '#ffa59e', '#f4777f', '#dd4c65', '#be214d', '#93003a'])
    )   

)
```


Vamos a enfocarnos en la Tienda_A para observar a detalle el comportamiento.

###Enfocando los datos

```python
filter_list = ['Tienda_A', 'Tienda_E']
filter_df = rpt_sales_op[rpt_sales_op['Store'].isin(filter_list)]
alt.Chart(filter_df).mark_point().encode(
    x='Op_MonthYear',
    y='Sales',
    size='AvgOP',
    color= alt.Color('Store', title='Tiendas',
              scale = alt.Scale(range=['#003790', '#be214d', '#93003a'])
    )   

)
```
